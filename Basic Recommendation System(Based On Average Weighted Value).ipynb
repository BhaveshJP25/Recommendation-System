# Recommendation System Based On Avergae Weighted Values

```python
import pandas as pd
import numpy as np
```

```python
c = pd.read_csv('https://raw.githubusercontent.com/BhaveshJP25/Dataset/1ad366a8aa11b64a24af245cba2b47a07fda6f53/Tmdb%20Movie%20Metadata/tmdb_5000_credits.csv')
```

```python
m = pd.read_csv('https://raw.githubusercontent.com/BhaveshJP25/Dataset/1ad366a8aa11b64a24af245cba2b47a07fda6f53/Tmdb%20Movie%20Metadata/tmdb_5000_movies.csv')
```

```python
credits = c.copy()
credits.head()
```

```python
movies_df = m.copy()
movies_df.head()
```

```python
print("Credits :\t\t", credits.shape, "\nMovies Dataframe :\t", movies_df.shape)
```

```python
movies_df_merged = pd.merge(movies_df, credits, how = 'inner', left_on = 'id', right_on = 'movie_id')
movies_df_merged.head()
# movies_df_merged.shape
```

```python
drop_columns = ['homepage', 'title_x', 'title_y', 'status', 'production_countries', 'movie_id']
movies_cleaned_df = movies_df_merged.drop(columns = drop_columns)
movies_cleaned_df.head()
```

```python
movies_cleaned_df.info()
```

# Using Weighted Average For Each Movie's Average Rating


##               W = (Rv + Cm) / (v + m)

####                   where
###                 W = Weighted rating
####                   R = Average for the movie as a number from 0 to 10 (mean) => Rating
####                   v = Number of votes for the movie => Votes
####                   m = Minimum votes required to be listed in Top 250 (Currently 3000)
####                   C = Mean votes across the whole report (currently 6.9)

```python
v = movies_cleaned_df['vote_count']
R = movies_cleaned_df['vote_average']
C = movies_cleaned_df['vote_average'].mean()
m = movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
movies_cleaned_df['weighted_average'] = ((R*v) + (C*m)) / (v+m)
movies_cleaned_df.head()
```

```python
movie_sorted_ranking = movies_cleaned_df.sort_values('weighted_average', ascending = False)
movie_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook

weight_average = movie_sorted_ranking.sort_values('weighted_average', ascending = False)
plt.figure(figsize = (15, 6))
plt.xlim(4, 10)
ax = sns.barplot(x = 'weighted_average', y = 'original_title', data = weight_average.head(10))
plt.title('Movie Rating By Average Votes', weight = 'bold')
plt.xlabel('Weighted Average Score', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.tight_layout()
plt.savefig('best_movies.png')
```

```python
popularity = movie_sorted_ranking.sort_values('popularity', ascending = False)
plt.figure(figsize = (12, 6))
ax = sns.barplot(x = 'popularity', y = 'original_title', data = popularity.head(10))
plt.title('Movie Popular By Votes', weight = 'bold')
plt.xlabel('Score Of Popularity', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.tight_layout()
plt.savefig('best_popular_movies.png')
```

## Recommendation Based On Scaled Weighted Average And Popularity Score ( Priority Is Given 50% to Both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()
movies_scaled_df = scaling.fit_transform(movies_cleaned_df[['weighted_average', 'popularity']])
movie_normalized_df = pd.DataFrame(movies_scaled_df, columns = ['weighted_average', 'popularity'])
movie_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weight_average', 'normalized_popularity']] = movie_normalized_df
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_weight_average'] * 0.5 + movies_cleaned_df['normalized_popularity'] * 0.5
movies_scored_df = movies_cleaned_df.sort_values(['score'], ascending = False)
movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
scored_df = movies_cleaned_df.sort_values('score', ascending = False)
plt.figure(figsize = (16, 6))
ax = sns.barplot(x = 'score', y = 'original_title', data = scored_df.head(10))
plt.title('Best Rated And Most Popular', weight = 'bold')
plt.xlabel('Score', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.tight_layout()
plt.savefig('scored_movies.png')
```
